```python
def add_text(stack, new_text):
    stack.append(new_text) #using append as push 
    print(f"Text added: '{new_text}'")
```

```python
def undo(stack):
    if stack:
        removed_text = stack.pop() #pop is pop
        print(f"Removed: '{removed_text}'")
    else:
        print("Nothing")
```

```python
def show_text(stack):
    print("Text Content:")
    for text in stack:
        print(text)
```

```python
def main():
    text_stack = []
    running = True

    while running:
        print("Menu")
        print("1. Add")
        print("2. Undo Last")
        print("3. Current Text")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            text = input("Enter: ")
            add_text(text_stack, text)

        elif choice == "2":
            undo(text_stack)

        elif choice == "3":
            show_text(text_stack)

        elif choice == "4":
            running = False
            print("Exiting Text Editor...")

        else:
            print("Invalid.")

if __name__ == "__main__":#checks correct name-o-logic
    main()
```

```python
class TextIterator:
    def __init__(self, data):
        self.data = data
        self.pos = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.pos >= len(self.data):
            raise StopIteration
        value = self.data[self.pos]
        self.pos += 1
        return value
```

```python
class Editor:
    def __init__(self):
        self.text = []
        self.undo_data = []

    def add_text(self, *lines):
        for line in lines:
            self.undo_data.append(self.text.copy())
            self.text.append(line)

    def undo(self):
        try:
            self.text = self.undo_data.pop()
        except IndexError:
            print("Nothing to undo")

    def show_generator(self):
        for line in self.text:
            yield line

    def show_iterator(self):
        data_iter = TextIterator(self.text)
        for line in data_iter:
            print(line)
```

```python


def main():
    editor = Editor()

    while True:
        print("\n1 Add text")
        print("2 Undo")
        print("3 Show using generator")
        print("4 Show using iterator")
        print("5 Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            line = input("Enter text: ")
            editor.add(line)

        elif choice == "2":
            editor.undo()

        elif choice == "3":
            for item in editor.show_generator():
                print(item)

        elif choice == "4":
            editor.show_iterator()

        elif choice == "5":
            break

        else:
            print("Invalid choice")


if __name__ == "__main__":
    main()
```

```python
class TopTen:

    def __init__(self):
        self.num = 1

    def __iter__(self):
        return self

    def __next__(self):

        if self.num <= 10:
            val = self.num
            self.num += 1

            return val
        else:
            raise StopIteration


values = TopTen()

print(next(values))

for i in values:
    print(i)
```

```python
from functools import reduce

nums = [3,2,6,8,4,6,2,9]

evens = list(filter(lambda n : n%2==0,nums))

doubles = list(map(lambda n : n*2,evens))
print(doubles)

sum = reduce(lambda a,b : a+b,doubles)

print(sum)
```
