```python
class FoodItem:
   
    def __init__(self, name: str, price: float):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} - ₹{self.price}"
```

```python
class Customer:

    def __init__(self, name: str):
        self.name = name
        self.cart = []

    def add_to_cart(self, food_item: FoodItem):
        self.cart.append(food_item)

    def calculate_total(self) -> float:
        total = 0
        for item in self.cart:
            total += item.price
        return total
```

```python
class Order:

    def __init__(self, customer: Customer):
        self.customer = customer

    def place_order(self):
        if not self.customer.cart:
            raise ValueError("Cart is empty. Cannot place order.")

        print(f"\nOrder placed by {self.customer.name}")
        print("Items ordered:")

        for item in self.customer.cart:
            print(item)

        total_amount = self.customer.calculate_total()
        print(f"Total Amount: ₹{total_amount}")
```

```python
def main():
    try:
        # Create food 
        pizza = FoodItem("Pizza", 250)
        burger = FoodItem("Burger", 150)
        fries = FoodItem("Fries", 100)

        #  customer
        customer = Customer("Manas")

        # Add 
        customer.add_to_cart(pizza)
        customer.add_to_cart(burger)
        customer.add_to_cart(fries)

        #  order
        order = Order(customer)
        order.place_order()

    except ValueError as error:
        print(f"Error: {error}")


if __name__ == "__main__":
    main()
```

```python
from abc import ABC, abstractmethod


class Product(ABC):
    product_count = 0  # Class variable
    
#constructor 
    def __init__(self, name: str, price: float, quantity: int):
        self.name = name
        self.price = price
        self.quantity = quantity
        

    @abstractmethod
    def calculate_discount(self) -> float:
        pass
#instance maethod
    def display_info(self) -> None:
        print(f"{self.name} | Price: ${self.price} | Quantity: {self.quantity}")
        
#dunder

    def __add__(self, other) -> int:
        if self.name == other.name:
            return self.quantity + other.quantity
        return self.quantity + other.quantity

    #class m 
    @classmethod
    def total_products(cls) -> int:
        return cls.product_count

    #static m
    @staticmethod
    def validate_price(price: float) -> bool:
        return price >= 0


#Inheritance 
class Electronics(Product):

    def calculate_discount(self) -> float:
        return self.price * 0.1


class Clothes(Product):

    def calculate_discount(self) -> float:
        return self.price * 0.2


```

```python

def main() -> None:
    
   
    laptop = Electronics("Laptop", 1000, 2)
    tshirt = Clothes("T-Shirt", 50, 5)

    laptop.display_info() 
    tshirt.display_info()  


    print("Laptop discount:", laptop.calculate_discount())
    print("T-Shirt discount:", tshirt.calculate_discount())

    
    print("Laptop price valid?", Product.validate_price(laptop.price))

 
    combined_qty = laptop + Electronics("Laptop", 1000, 3)
    print("Total Laptop quantity combined:", combined_qty)




if __name__ == "__main__":
    main()
```

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")
```

```python
def calculate(num1, num2, operation):
    if operation == '+':
        return num1 + num2
    elif operation == '-':
        return num1 - num2
    elif operation == '*':
        return num1 * num2
    elif operation == '/':
        return num1 / num2
    else:
        raise ValueError("Invalid")
```

```python
while True:
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        operation = input("Enter operation : ")

        result = calculate(num1, num2, operation)
        print("Result:", result)

    except ValueError as ve:
        print("Error:", ve)

    except ZeroDivisionError:
        print("Error:not divide by zero!")

    except Exception as e:
        print("Unexpected :", e)

    t = input("Do you want to work (yes/no): ")
    if t != 'yes':
        print("exiting...")
        break
```
