```python
products = []

def add_product(name, price, quantity):
    product = {
        "name": name,
        "price": price,
        "quantity": quantity
    }
    products.append(product)
    print(f"Product '{name}' added")
```

```python
def sell_product(name, sell_qty):
    for product in products:
        if product["name"] == name:
            if product["quantity"] >= sell_qty:
                product["quantity"] -= sell_qty
                print(f"Sold {sell_qty} {name}")
            else:
                print("Not enough stock")
            return
    print("Product not found")
```

```python
def show_products():
    if not products:
        print("No products available")
    for product in products:
        print(f"{product['name']} - Price: {product['price']}, Quantity: {product['quantity']}")
```

```python
add_product("Apple", 10, 50)
add_product("Banana", 5, 30)
sell_product("Apple", 20)
sell_product("Orange", 5)
show_products()
```

```python
def fact(n):

    f = 1

    for i in range(1,n+1):
        f = f * i
    return f


x = 4

result = fact(x)


print(result)
```

```python
def fib(n):

    a = 0
    b = 1

    if n == 1:
        print(a)

    else:
        print(a)
        print(b)

    for i in range(2,n):
        c = a + b
        a = b
        b = c
        print(a+b)

fib(10)
```
