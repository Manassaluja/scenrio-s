```python
students = []

def add_student():
    try:
        name = input("Enter name: ")
        marks = float(input("Enter student marks: "))
        student = {"name": name, "marks": marks}
        students.append(student)
        print("Student added .")
    except ValueError:
        print("Error: marks must be a number.")

def update_marks():
    try:
        name = input("Enter name to update marks: ")
        for student in students:
            if student["name"].lower() == name.lower():
                new_marks = float(input("Enter new marks: "))
                student["marks"] = new_marks
                print("Marks updated.")
                return
        raise ValueError("not found.")
    except ValueError as ve:
        print("Error:", ve)
```

```python
def view_students():
    if not students:
        print("No one found.")
    else:
        for idx, student in enumerate(students, start=1):
            print(f"{idx}. Name: {student['name']}, Marks: {student['marks']}")
```

```python
def search_student():
    try:
        name = input("Enter student name to search: ")
        for student in students:
            if student["name"].lower() == name.lower():
                print(f"Found: Name: {student['name']}, Marks: {student['marks']}")
                return
        raise ValueError("Student not found.")
    except ValueError as ve:
        print("Error:", ve)

while True:
    print("\n--- Student Management System ---")
    print("1. Add Student")
    print("2. Update Marks")
    print("3. View All Students")
    print("4. Search Student")
    print("5. Exit")

    try:
        choice = int(input("Enter your choice (1-5): "))
        if choice == 1:
            add_student()
        elif choice == 2:
            update_marks()
        elif choice == 3:
            view_students()
        elif choice == 4:
            search_student()
        elif choice == 5:
            print("Exiting system...")
            break
        else:
            print("Invalid choice. Enter 1-5.")
    except ValueError:
        print("Error: Please enter a valid number.")
```

```python
from abc import ABC, abstractmethod


class StudentOperation(ABC):

    @abstractmethod
    def add_student(self):
        pass

    @abstractmethod
    def update_marks(self):
        pass

    @abstractmethod
    def view_students(self):
        pass

    @abstractmethod
    def search_student(self):
        pass
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
```

```python
class Studentm(StudentOperation):

    def __init__(self):
        self.students = []

    def add_student(self):
        try:
            name = input("Enter name: ")
            marks = float(input("Enter marks: "))

            student = Student(name, marks)
            self.students.append(student)
            print("Student added .")

        except ValueError:
            print("Error: Invalid marks.")

    def update_marks(self):
        try:
            name = input("Enter name to update marks: ")

            for student in self.students:
                if student.name.lower() == name.lower():
                    student.marks = float(input("Enter new marks: "))
                    print("Marks updated successfully.")
                    return

            raise ValueError("not found.")

        except ValueError as ve:
            print("Error:", ve)

    def view_students(self):
        if not self.students:
            print("No students available.")
            return

        for index, student in enumerate(self.students, start=1):
            print(
                f"{index}. Name: {student.name}, "
                f"Marks: {student.marks}"
            )

    def search_student(self):
        try:
            name = input("Enter student name to search: ")

            for student in self.students:
                if student.name.lower() == name.lower():
                    print(
                        f"Found â†’ Name: {student.name},"
                        f"Marks: {student.marks}"
                    )
                    return

            raise ValueError("not found.")

        except ValueError as ve:
            print("Error:", ve)


system = Studentm()

while True:
    print("\n--- Student Management System ---")
    print("1. Add Student")
    print("2. Update Marks")
    print("3. View Students")
    print("4. Search Student")
    print("5. Exit")

    try:
        choice = int(input("Enter choice (1-5): "))

        if choice == 1:
            system.add_student()
        elif choice == 2:
            system.update_marks()
        elif choice == 3:
            system.view_students()
        elif choice == 4:
            system.search_student()
        elif choice == 5:
            print("Exiting system...")
            break
        else:
            print("Invalid choice.")

    except ValueError:
        print("Error: Please enter a valid number.")
```
